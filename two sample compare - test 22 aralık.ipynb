```python
import random
import numpy as np
import matplotlib.pyplot as plt 
```

```python
numbers=[]
size=10000
for i in range(size):
    number=random.gauss(50,15)
    numbers.append(number)
```

```python
def generate_population_with_size_mean_and_sd(size=10000,mean=50,SD=15):
    """ 
    size=10000,mean=50,SD=15

    """

    numbers=[]
    # size=10000
    for i in range(size):
        number=random.gauss(mean,SD)
        numbers.append(number)
    return numbers
pop_1=generate_population_with_size_mean_and_sd()
np.mean(pop_1),np.std(pop_1),len(pop_1)
```

```python
def get_sample_from_a_population(size=100,pop_1=[]):
    if len(pop_1)==0:
        numbers=generate_population_with_size_mean_and_sd(10000,50,15)
    else:
        numbers=pop_1
    size_of_sample=size
    sample=[]
    for i in range(size_of_sample):
        number=random.choice(numbers)
        sample.append(number)
    return sample


pop_1=generate_population_with_size_mean_and_sd(20000,30,20)
sample_1=get_sample_from_a_population(90,pop_1)
np.mean(sample_1),np.std(sample_1),len(sample_1)
```

```python
def get_sample_statistics(sample_1):
    return np.mean(sample_1),np.std(sample_1),len(sample_1)

get_sample_statistics(sample_1)
```

```python
def get_two_sample_difference():
    #
    population_size=10000
    # for sample 1:
    avg_1=285
    sd_1=37
    sample_size_1=1000
    
    population_1=generate_population_with_size_mean_and_sd(population_size,avg_1,sd_1)   
    sample_1=get_sample_from_a_population(sample_size_1,population_1)
    # sample statistics
    mean_1,sd_1,n_1=get_sample_statistics(sample_1)

    # for sample 2:
    avg_2=290
    sd_2=40
    sample_size_2=1000
    
    population_2=generate_population_with_size_mean_and_sd(population_size,avg_2,sd_2)   
    sample_2=get_sample_from_a_population(sample_size_2,population_2)
    # sample statistics
    mean_2,sd_2,n_2=get_sample_statistics(sample_2)

    if 1==0:

        print(mean_1,sd_1,n_1)

        print(mean_2,sd_2,n_2)

    return mean_1-mean_2
get_two_sample_difference()
```

```python
two_sample_differences=[]
for i in range(100):
    r=get_two_sample_difference()
    two_sample_differences.append(r)

a=np.average(two_sample_differences)
b=np.std(two_sample_differences)

a,b
```

```python
#  plot the values named ;   two_sample_differences
y=two_sample_differences
x=[i for i in range(len(two_sample_differences))]
# plt.plot(two_sample_differences)
points=zip(two_sample_differences)
plt.scatter(x,y)
a=np.mean(two_sample_differences)
b=np.std(two_sample_differences)
title_plt= "two sample difference mean "  + str(a) +"  std "+ str(b)
plt.title(title_plt)
plt.show()
```
